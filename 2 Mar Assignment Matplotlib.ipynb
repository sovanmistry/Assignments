Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
Note: Also add title, xlabel, and ylabel to the plot.



```python

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

Scatter PLot: Scatter plots are the graphs that present the relationship between two variables in a data-set. 

```python
import matplotlib.pyplot as plt

plt.scatter(x, y )
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("TITLE")
```

subplots: 

creates a figure and a grid of subplots with a single call, while providing reasonable control over how the individual plots are created. 

```python
import numpy as np
import matplotlib.pyplot as plt


fig, axes = plt.subplots(nrows = 4, ncols=1)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
axes[0].plot(x,y )


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
axes[1].plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
axes[2].plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
axes[3].plot(x,y)
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])


Bar Plot: A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights 
that is proportional to the values which they represent. 
The bar plots can be plotted horizontally or vertically

A bar chart describes the comparisons between the discrete categories.
One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt

plt.bar(x = company, height= profit)
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

box Plot:
    
A Box Plot is also known as Whisker plot is created to display the summary of the set of data values 
having properties like minimum, first quartile, median, third quartile and maximum. 

In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. 
Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.
    

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
```

```python
plt.boxplot( box1)
```

```python
plt.boxplot( box2)
```

```python
plt.boxplot( data)
```
